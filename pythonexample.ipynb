```python
fullLine = ''
comma = ','
seperator = '|'
dataCats = []
dataNums = []

dataTitle = input('Enter a title for the data:\n')
print('You entered: %s\n' % dataTitle)

col1Header = input('Enter the column 1 header:\n')
print('You entered: %s\n' % col1Header)
col2Header = input('Enter the column 2 header:\n')
print('You entered: %s\n' % col2Header)

while fullLine != '-1':
   fullLine = input('Enter a data point (-1 to stop input):\n')
   
   while fullLine.find(comma) == -1 and fullLine != '-1':
     print('Error: No comma in string.\n')
     fullLine = input('Enter a data point (-1 to stop input):\n')
   
   if fullLine != '-1':
      my_tokens = fullLine.split(comma)
   
      # Too many tokens = too many commas
      while len(my_tokens) > 2 and fullLine != '-1':
         print('Error: Too many commas in input.\n')
         fullLine = input('Enter a data point (-1 to stop input):\n')
         
         if fullLine != '-1':
            my_tokens = fullLine.split(comma)
   
   if fullLine != '-1':
      while not my_tokens[1][1:].isdigit() and fullLine != '-1':
         print('Error: Comma not followed by an integer.\n')
         fullLine = input('Enter a data point (-1 to stop input):\n')
         
         if fullLine != '-1':
            my_tokens = fullLine.split(comma)
   
   if fullLine != '-1':   
      dataString = my_tokens[0];
      dataInt = int(my_tokens[1][1:])
      print('Data string:', dataString)
      print ('Data integer: %d\n' % dataInt)
   
      dataCats.append(dataString)
      dataNums.append(dataInt)

print()
print('%33s' % dataTitle)
print('%-20s|%23s' % (col1Header, col2Header))
print('--------------------------------------------')
for i in range (len(dataCats)):
   print('%-20s|%23s' % (dataCats[i], str(dataNums[i])))


for i in range (len(dataCats)):
   print('\n%20s' % dataCats[i], end = " ")
   for j in range(dataNums[i]):
      print('*', end = "")
   
print()
```
